```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('attendance.csv')
```

```python
df.head()
```

```python
df_melt=df.melt(id_vars=['Unnamed: 0'], var_name='date', value_name='attendance')
df_melt.head()

```

```python
df_melt['attendance']=df_melt['attendance'].replace(['A', 'P', 'H', 'T'],[0, 1, 0.5, 0.9])
```

```python
df_melt.head()
```

```python
df_melt = df_melt.groupby(['Unnamed: 0']).mean()
```

```python
df_melt.head()
```

```python
cakes=pd.read_csv('cake_recipes.csv')
```

```python
cakes=cakes.drop(columns=['recipe:position'])
```

```python
cakes=cakes.melt(id_vars=['recipe', 'position'], var_name='temperature', value_name='rating')
cakes
```
