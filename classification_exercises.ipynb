```python
def prep_titanic(titanic):
    titanic=titanic.drop(columns=['age', 'deck', 'class'])
    titanic=titanic.drop(columns=['embarked'])
    dummy_df=pd.get_dummies(titanic[['sex', 'embark_town']], drop_first=True)
    titanic=pd.concat([titanic, dummy_df], axis=1)
    titanic.drop(columns=['sex', 'embark_town', 'Unnamed: 0'])
    return titanic
    
    
```
